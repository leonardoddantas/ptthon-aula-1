```python
idade = '30'
print(idade)
print(type(idade))

idade_int = int(idade)
print(idade_int)
print(type(idade_int))

idade_float = float(idade_int)
print(idade_float)
print(type(idade_float))

idade_str = str(idade_float)
print(idade_str)
print(type(idade_str))
```

# Operações matemátcas

```python
a, b = 5, 10
print(a + b) #soma
print(a - b) #subtração
print(a * b) #multiplicação
print(a / b) #divisão
print(a // b) #divisão inteira
print(a % b) #módulo
print(a ** b) #exponenciação
print(a == b) #comparação de igual
print(a != b) #soma
print(a > b) #soma
print(a < b) #soma
print(a <= b) #soma
print(a >= b) #soma
```

```python
#operadores lógicos (AND, OR, NOT)
maior_de_idade = True
tem_cnh = False
print(maior_de_idade and tem_cnh)
```

```python
print(abs(-5)) #valor absoluto
print(round(3.14159, 2)) 
print(max(1, 5, 3))
print(min(1, 5, 3))
```
